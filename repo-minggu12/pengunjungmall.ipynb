```python
# Import Library yang digunakan 
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans
```

```python
# Menyiapkan data dan memanggil dataset
dataset = pd.read_csv('pengunjung_Mall.csv')
dataset.keys()
```

```python
# Menampilkan 5 baris data pertama dataset
dataku = pd.DataFrame(dataset)
dataku.head()
```

```python
# Konversi Dataset ke Data Array
X = np.asarray(dataset)
print(X)
```

```python
# Menampilkan data dalam bentuk scatter plot
plt.scatter(X[:,2], X[:,4], label='True Position')
plt.xlabel("Usia")
plt.ylabel("Pengeluaran_USD")
plt.title("Grafik Data Usia dan Pengeluaran USD pada Pengunjung Mall")
plt.show()
```

```python
# Mengaktifkan K-Means dengan jumlah K=2
kmeans = KMeans(n_clusters=2)
kmeans.fit(X)
```

```python
print(kmeans.cluster_centers_)
```

```python
# Plot Data Point
# Memvisualisasikan Hasil Klasterisasi Data Konsumen
plt.scatter(X[:,2], X[:,4], c=kmeans.labels_, cmap='rainbow')
plt.xlabel("Usia")
plt.ylabel("Pengeluaran_USD")
plt.title("Grafik Hasil Klasterisasi Data Usia dan Pengeluaran USD pada Pengunjung Mall")
plt.show()
```

```python
# Plot Data Point
# Memvisualisasikan Hasil Klasterisasi dengan centroid dari masing masing klaster
plt.scatter(X[:,2], X[:,4], c=kmeans.labels_, cmap='rainbow')
plt.scatter(kmeans.cluster_centers_[:,2], kmeans.cluster_centers_[:,4], color='black')
plt.xlabel("Usia")
plt.ylabel("Pengeluaran_USD")
plt.title("Grafik Hasil Klasterisasi Data Usia dan Pengeluaran USD pada Pengunjung Mall")
plt.show()
```
