```python
# Import Library yang digunakan 
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans
```

```python
# Menyiapkan data dan memanggil dataset
dataset = pd.read_csv('pendapatan.csv')
dataset.keys()
```

```python
# Menampilkan 5 baris data pertama dataset
dataku = pd.DataFrame(dataset)
dataku.head()
```

```python
# Konversi Dataset ke Data Array
X = np.asarray(dataset)
print(X)
```

```python
# Menampilkan data dalam bentuk scatter plot
plt.scatter(X[:,0], X[:,1], label='True Position')
plt.xlabel("tahun")
plt.ylabel("perkapita_jakarta")
plt.title("Grafik Pendapatan Perkapita Jakarta")
plt.show()
```

```python
# Mengaktifkan K-Means dengan jumlah K=1
kmeans = KMeans(n_clusters=1)
kmeans.fit(X)
```

```python
# Menampilkan nilai Centroid yang digenerate oleh algoritma
print(kmeans.cluster_centers_)
```

```python
# Plot Data Point
# Memvisualisasikan Hasil Klasterisasi Data Konsumen
plt.scatter(X[:,0], X[:,1], c=kmeans.labels_, cmap='rainbow')
plt.xlabel("tahun")
plt.ylabel("perkapita_jakarta")
plt.title("Grafik Hasil Klasterisasi Data Tahun dan Pendapatan Perkapita Jakarta")
plt.show()
```

```python
# Plot Data Point
# Memvisualisasikan Hasil Klasterisasi dengan centroid dari masing masing klaster
plt.scatter(X[:,0], X[:,1], c=kmeans.labels_, cmap='rainbow')
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], color='black')
plt.xlabel("tahun")
plt.ylabel("perkapita_jakarta")
plt.title("Grafik Hasil Klasterisasi Data Tahun dan Pendapatan Perkapita Jakarta dengan Centroid")
plt.show()
```
